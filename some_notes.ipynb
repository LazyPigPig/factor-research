```python
def normalize_df(df):
    """按行进行Z-Score归一化，处理标准差为0的情况"""
    row_means = df.mean(axis=1)
    row_stds = df.std(axis=1)
    row_stds[row_stds == 0] = 1  # 处理标准差为0的情况
    return df.sub(row_means, axis=0).div(row_stds, axis=0)
```
