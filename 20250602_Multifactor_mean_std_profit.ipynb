```python
class CryptoFourFactorModel:
    def __init__(self, data):
        """
        初始化四因子模型
        :param data: DataFrame, 包含以下列: ['date', 'symbol', 'open', 'high', 'low', 'close', 'volume']
        """
        self.data = data.copy()
        self.factors = None
        self.portfolio_returns = None
        self.portfolio_stats = None
        self.model_results = None
        self._preprocess_data()
    
    def _preprocess_data(self):
        """数据预处理：计算收益率和特征"""
        print("Preprocessing data...")
        # 确保日期格式正确
        self.data['date'] = pd.to_datetime(self.data['date'])
        self.data.sort_values(['symbol', 'date'], inplace=True)
        
        # 计算日收益率
        self.data['return'] = self.data.groupby('symbol')['close'].pct_change()
        
        # 计算波动率 (过去30天标准差)
        self.data['volatility'] = self.data.groupby('symbol')['return'].transform(
            lambda x: x.rolling(window=30, min_periods=15).std())
        
        # 计算动量 (过去30天累计收益率)
        self.data['momentum'] = self.data.groupby('symbol')['return'].transform(
            lambda x: x.rolling(window=30, min_periods=15).apply(lambda y: (1+y).prod()-1))
        
        # 计算规模 (过去30天平均交易量)
        self.data['volume_avg'] = self.data.groupby('symbol')['volume'].transform(
            lambda x: x.rolling(window=30, min_periods=15).mean())
        
        # 计算市场收益率 (所有币种等权重平均)
        market_return = self.data.groupby('date')['return'].mean().reset_index()
        market_return.columns = ['date', 'market_return']
        self.data = pd.merge(self.data, market_return, on='date')
        
        # 计算因子值
        self._calculate_factors()
    
    def _calculate_factors(self):
        """计算四因子"""
        print("Calculating factors...")
        
        # 市场因子 (Market Factor)
        market_factor = self.data.groupby('date')['return'].mean().reset_index()
        market_factor.columns = ['date', 'MKT']
        
        # 规模因子 (Size Factor)
        # 每天根据规模(交易量)将币种分成大小两组，计算小规模组和大规模组的平均收益率之差
        size_factor = []
        for date, group in tqdm(self.data.groupby('date'), desc="Calculating size factor"):
            if len(group) < 10:  # 至少需要10个币种才能计算
                continue
            group = group.sort_values('volume_avg')
            small_size = group.iloc[:len(group)//2]
            big_size = group.iloc[len(group)//2:]
            sbm = small_size['return'].mean() - big_size['return'].mean()
            size_factor.append({'date': date, 'SMB': sbm})
        size_factor = pd.DataFrame(size_factor)
        
        # 动量因子 (Momentum Factor)
        mom_factor = []
        for date, group in tqdm(self.data.groupby('date'), desc="Calculating momentum factor"):
            if len(group) < 10:
                continue
            group = group.sort_values('momentum')
            winner = group.iloc[-len(group)//3:]  # 前1/3为赢家
            loser = group.iloc[:len(group)//3]    # 后1/3为输家
            mom = winner['return'].mean() - loser['return'].mean()
            mom_factor.append({'date': date, 'MOM': mom})
        mom_factor = pd.DataFrame(mom_factor)
        
        # 波动率因子 (Volatility Factor)
        vol_factor = []
        for date, group in tqdm(self.data.groupby('date'), desc="Calculating volatility factor"):
            if len(group) < 10:
                continue
            group = group.sort_values('volatility')
            low_vol = group.iloc[:len(group)//3]   # 低波动率组
            high_vol = group.iloc[-len(group)//3:]  # 高波动率组
            vol = low_vol['return'].mean() - high_vol['return'].mean()
            vol_factor.append({'date': date, 'VOL': vol})
        vol_factor = pd.DataFrame(vol_factor)
        
        # 合并所有因子
        self.factors = market_factor.merge(size_factor, on='date', how='left')
        self.factors = self.factors.merge(mom_factor, on='date', how='left')
        self.factors = self.factors.merge(vol_factor, on='date', how='left')
        self.factors.set_index('date', inplace=True)
        self.factors.dropna(inplace=True)
        
        # 标准化因子
        scaler = StandardScaler()
        self.factors[['SMB', 'MOM', 'VOL']] = scaler.fit_transform(self.factors[['SMB', 'MOM', 'VOL']])
        
        # 添加常数项用于回归
        self.factors['const'] = 1
    
    def single_factor_test(self):
        """单因子检验：分析每个因子的表现"""
        print("Performing single factor tests...")
        
        # 计算每个因子的描述性统计
        factor_stats = self.factors[['MKT', 'SMB', 'MOM', 'VOL']].describe().T
        factor_stats['t-stat'] = factor_stats['mean'] / (factor_stats['std'] / np.sqrt(factor_stats['count']))
        
        # 绘制因子时间序列
        fig, axes = plt.subplots(4, 1, figsize=(14, 16))
        for i, factor in enumerate(['MKT', 'SMB', 'MOM', 'VOL']):
            self.factors[factor].plot(ax=axes[i], title=f'{factor} Factor Time Series')
            axes[i].set_ylabel('Factor Return')
        plt.tight_layout()
        plt.savefig('factor_time_series.png', dpi=300)
        
        # 绘制因子相关性热力图
        plt.figure(figsize=(10, 8))
        sns.heatmap(self.factors[['MKT', 'SMB', 'MOM', 'VOL']].corr(), 
                    annot=True, cmap='coolwarm', center=0)
        plt.title('Factor Correlation Matrix')
        plt.tight_layout()
        plt.savefig('factor_correlation.png', dpi=300)
        
        return factor_stats
    
    def run_multifactor_regression(self):
        """运行多因子回归分析"""
        print("Running multifactor regression...")
        
        # 准备数据
        data = self.data.copy()
        data = data[data['date'].isin(self.factors.index)]
        
        # 创建结果容器
        results = []
        
        # 对每个币种运行时间序列回归
        for symbol, symbol_data in tqdm(data.groupby('symbol'), desc="Running regressions"):
            if len(symbol_data) < 100:  # 至少需要100个观测值
                continue
            
            # 合并因子数据
            merged = symbol_data.merge(self.factors.reset_index(), on='date')
            
            # 因变量：币种超额收益率
            y = merged['return']
            
            # 自变量：四因子
            X = merged[['const', 'MKT', 'SMB', 'MOM', 'VOL']]
            
            try:
                # 运行OLS回归
                model = sm.OLS(y, X, missing='drop').fit()
                
                # 存储结果
                res = {
                    'symbol': symbol,
                    'alpha': model.params['const'],
                    'beta_mkt': model.params['MKT'],
                    'beta_smb': model.params['SMB'],
                    'beta_mom': model.params['MOM'],
                    'beta_vol': model.params['VOL'],
                    'r_squared': model.rsquared,
                    't_alpha': model.tvalues['const'],
                    't_mkt': model.tvalues['MKT'],
                    't_smb': model.tvalues['SMB'],
                    't_mom': model.tvalues['MOM'],
                    't_vol': model.tvalues['VOL'],
                    'nobs': model.nobs
                }
                results.append(res)
            except:
                continue
        
        self.model_results = pd.DataFrame(results)
        
        # 分析回归结果
        self._analyze_regression_results()
        
        return self.model_results
    
    def _analyze_regression_results(self):
        """分析回归结果"""
        print("Analyzing regression results...")
        
        # 整体平均因子暴露
        avg_exposure = self.model_results[['beta_mkt', 'beta_smb', 'beta_mom', 'beta_vol']].mean()
        t_stats = self.model_results[['t_mkt', 't_smb', 't_mom', 't_vol']].mean()
        
        exposure_stats = pd.DataFrame({
            'Average Beta': avg_exposure,
            'Average t-stat': t_stats
        })
        
        # 绘制因子暴露分布
        fig, axes = plt.subplots(2, 2, figsize=(14, 10))
        axes = axes.flatten()
        
        for i, factor in enumerate(['beta_mkt', 'beta_smb', 'beta_mom', 'beta_vol']):
            sns.histplot(self.model_results[factor], kde=True, ax=axes[i])
            axes[i].set_title(f'{factor[5:].upper()} Beta Distribution')
            axes[i].set_xlabel('Beta Value')
            axes[i].axvline(self.model_results[factor].mean(), color='r', linestyle='--')
        
        plt.tight_layout()
        plt.savefig('beta_distribution.png', dpi=300)
        
        # 绘制R²分布
        plt.figure(figsize=(10, 6))
        sns.histplot(self.model_results['r_squared'], kde=True)
        plt.axvline(self.model_results['r_squared'].mean(), color='r', linestyle='--')
        plt.title('Distribution of R-squared Values')
        plt.xlabel('R-squared')
        plt.tight_layout()
        plt.savefig('rsquared_distribution.png', dpi=300)
        
        return exposure_stats
    
    def build_min_variance_portfolio(self, target_return=0.001, top_n=50, min_weight=0.01, max_weight=0.15, 
                                    lookback=252, risk_free_rate=0.0):
        """
        构建目标收益率下最小方差的投资组合
        
        :param target_return: 目标日收益率
        :param top_n: 考虑的前N个币种
        :param min_weight: 最小权重限制
        :param max_weight: 最大权重限制
        :param lookback: 用于计算协方差矩阵的回看天数
        :param risk_free_rate: 无风险利率
        """
        print(f"Building min variance portfolio with target return: {target_return*100:.2f}% per day...")
        
        # 获取最近的日期范围
        end_date = self.data['date'].max()
        start_date = end_date - pd.Timedelta(days=lookback)
        recent_data = self.data[self.data['date'] >= start_date]
        
        # 选择流动性最好的top_n个币种
        liquidity = recent_data.groupby('symbol')['volume'].mean().sort_values(ascending=False)
        top_symbols = liquidity.head(top_n).index.tolist()
        candidate_data = recent_data[recent_data['symbol'].isin(top_symbols)]
        
        # 创建收益率矩阵 (日期 × 币种)
        returns_matrix = candidate_data.pivot(index='date', columns='symbol', values='return')
        returns_matrix = returns_matrix.dropna(axis=1)  # 删除全为NaN的列
        
        # 计算预期收益率和协方差矩阵
        expected_returns = returns_matrix.mean()
        # 使用Ledoit-Wolf收缩估计协方差矩阵
        cov_matrix = LedoitWolf().fit(returns_matrix).covariance_
        
        # 币种数量
        n_assets = len(expected_returns)
        
        # 定义优化问题
        weights = cp.Variable(n_assets)
        
        # 目标函数: 最小化方差
        portfolio_variance = cp.quad_form(weights, cov_matrix)
        
        # 约束条件
        constraints = [
            cp.sum(weights) == 1,  # 权重和为1
            expected_returns.values @ weights >= target_return,  # 目标收益率约束
            weights >= min_weight,  # 最小权重
            weights <= max_weight   # 最大权重
        ]
        
        # 优化问题
        problem = cp.Problem(cp.Minimize(portfolio_variance), constraints)
        
        # 求解
        try:
            problem.solve(solver=cp.ECOS)
        except Exception as e:
            print(f"Optimization failed: {e}")
            # 如果优化失败，使用等权重组合
            equal_weights = np.ones(n_assets) / n_assets
            weights_value = equal_weights
        else:
            if problem.status not in ["optimal", "optimal_inaccurate"]:
                print(f"Optimization status: {problem.status}")
                equal_weights = np.ones(n_assets) / n_assets
                weights_value = equal_weights
            else:
                weights_value = weights.value
        
        # 处理负权重（理论上不应该出现，但以防万一）
        weights_value = np.maximum(weights_value, 0)
        weights_value /= np.sum(weights_value)  # 重新归一化
        
        # 创建权重DataFrame
        weights_df = pd.DataFrame({
            'symbol': expected_returns.index,
            'weight': weights_value
        })
        
        # 过滤掉权重过小的币种
        weights_df = weights_df[weights_df['weight'] > 0.005].sort_values('weight', ascending=False)
        
        # 计算组合的预期收益率和波动率
        portfolio_return = expected_returns.dot(weights_df.set_index('symbol')['weight'])
        portfolio_volatility = np.sqrt(weights_df['weight'].T @ cov_matrix.loc[weights_df['symbol'], weights_df['symbol']] @ weights_df['weight'])
        
        # 计算夏普比率
        sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility * np.sqrt(252)
        
        # 计算组合的历史表现
        portfolio_history = self._calculate_portfolio_history(weights_df, lookback)
        
        # 计算最大回撤
        max_drawdown = self._calculate_max_drawdown(portfolio_history)
        
        # 绘制有效前沿
        self._plot_efficient_frontier(expected_returns, cov_matrix, target_return, weights_df)
        
        return weights_df, portfolio_return, portfolio_volatility, sharpe_ratio, max_drawdown, portfolio_history
    
    def _calculate_portfolio_history(self, weights_df, lookback):
        """计算组合的历史表现"""
        # 获取权重币种
        symbols = weights_df['symbol'].tolist()
        weights = weights_df['weight'].values
        
        # 获取历史数据
        end_date = self.data['date'].max()
        start_date = end_date - pd.Timedelta(days=lookback)
        portfolio_data = self.data[(self.data['date'] >= start_date) & 
                                  (self.data['symbol'].isin(symbols))]
        
        # 创建收益率矩阵
        returns_matrix = portfolio_data.pivot(index='date', columns='symbol', values='return')
        returns_matrix = returns_matrix.fillna(0)  # 填充缺失值
        
        # 确保矩阵列顺序与权重一致
        returns_matrix = returns_matrix[weights_df['symbol']]
        
        # 计算每日组合收益率
        portfolio_returns = returns_matrix.dot(weights)
        
        # 计算累计收益
        cumulative_returns = (1 + portfolio_returns).cumprod() - 1
        
        return pd.DataFrame({
            'date': returns_matrix.index,
            'daily_return': portfolio_returns,
            'cumulative_return': cumulative_returns
        })
    
    def _calculate_max_drawdown(self, portfolio_history):
        """计算最大回撤"""
        cumulative = portfolio_history['cumulative_return'] + 1
        peak = cumulative.cummax()
        drawdown = (cumulative - peak) / peak
        return drawdown.min()
    
    def _plot_efficient_frontier(self, expected_returns, cov_matrix, target_return, weights_df):
        """绘制有效前沿和优化结果"""
        # 生成随机组合
        n_portfolios = 10000
        results = np.zeros((3, n_portfolios))
        
        for i in range(n_portfolios):
            # 生成随机权重
            weights = np.random.random(len(expected_returns))
            weights /= np.sum(weights)
            
            # 计算组合收益率和波动率
            portfolio_return = np.sum(expected_returns * weights)
            portfolio_volatility = np.sqrt(weights.T @ cov_matrix @ weights)
            
            # 存储结果
            results[0, i] = portfolio_return
            results[1, i] = portfolio_volatility
            results[2, i] = (portfolio_return - 0) / portfolio_volatility  # 夏普比率 (无风险利率=0)
        
        # 计算优化组合的指标
        opt_return = expected_returns.dot(weights_df.set_index('symbol')['weight'])
        opt_volatility = np.sqrt(weights_df['weight'].T @ cov_matrix.loc[weights_df['symbol'], weights_df['symbol']] @ weights_df['weight'])
        opt_sharpe = (opt_return - 0) / opt_volatility
        
        # 创建绘图
        plt.figure(figsize=(12, 8))
        
        # 绘制随机组合
        plt.scatter(results[1, :], results[0, :], 
                   c=results[2, :], cmap='viridis', 
                   marker='o', s=10, alpha=0.3)
        plt.colorbar(label='Sharpe Ratio')
        
        # 标记优化组合
        plt.scatter(opt_volatility, opt_return, 
                   color='r', s=100, marker='*', 
                   label=f'Optimized Portfolio (Return: {opt_return*100:.2f}%, Vol: {opt_volatility*100:.2f}%)')
        
        # 标记目标收益率线
        plt.axhline(y=target_return, color='b', linestyle='--', 
                   label=f'Target Return: {target_return*100:.2f}%')
        
        # 标记市场组合 (等权重)
        market_return = expected_returns.mean()
        market_volatility = np.sqrt(cov_matrix.values.mean() * len(expected_returns))
        plt.scatter(market_volatility, market_return, 
                   color='g', s=100, marker='s', 
                   label=f'Market Portfolio (Return: {market_return*100:.2f}%, Vol: {market_volatility*100:.2f}%)')
        
        # 设置图表
        plt.title('Efficient Frontier with Minimum Variance Portfolio')
        plt.xlabel('Daily Volatility')
        plt.ylabel('Daily Expected Return')
        plt.legend(loc='upper left')
        plt.grid(True)
        plt.tight_layout()
        plt.savefig('efficient_frontier.png', dpi=300)
        
        return plt
    
    def plot_portfolio_performance(self, portfolio_history):
        """绘制组合表现图"""
        plt.figure(figsize=(12, 8))
        
        # 绘制累计收益
        plt.subplot(2, 1, 1)
        plt.plot(portfolio_history['date'], portfolio_history['cumulative_return'], 
                label='Portfolio Cumulative Return')
        plt.title('Portfolio Cumulative Return')
        plt.ylabel('Cumulative Return')
        plt.legend()
        plt.grid(True)
        
        # 绘制每日收益
        plt.subplot(2, 1, 2)
        plt.bar(portfolio_history['date'], portfolio_history['daily_return'],
               color=np.where(portfolio_history['daily_return'] > 0, 'g', 'r'))
        plt.title('Portfolio Daily Returns')
        plt.ylabel('Daily Return')
        plt.xlabel('Date')
        plt.grid(True)
        
        plt.tight_layout()
        plt.savefig('portfolio_performance.png', dpi=300)
        
        return plt
```
